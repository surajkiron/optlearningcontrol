$ -e^{-(x-1)^2} \:where, \: x \in R $ 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return -np.exp(-(x - 1)**2)

# Generate x values
x_values = np.linspace(-3, 5, 400)

# Compute the corresponding y values
y_values = f(x_values)

def plot():
    # Compute the minimum and maximum y values
    y_min = np.min(y_values)
    y_max = np.max(y_values)

    # Plot the function
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, label=r'$-e^{-(x-1)^2}$')

    # Add horizontal lines at the minimum and maximum y values
    plt.axhline(y=y_min, color='r', linestyle='--', label=f'Min: {y_min:.2f}')
    plt.axhline(y=y_max, color='g', linestyle='--', label=f'Max: {y_max:.2f}')

    # Add title, labels, grid, and legend
    # plt.title('Plot of $-e^{-(x-1)^2}$ with Min/Max Horizontal Lines')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()

    # Show the plot
    plt.show()

# Call the plot function
plot() 
```

```python
# Necessary imports after reset
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the contour plot
def f_contour(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Generate x and y values for the contour plot
x_values = np.linspace(-2, 2, 400)
y_values = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x_values, y_values)

# Compute the corresponding z values (function values)
Z = f_contour(X, Y)

# Create the contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=1000, cmap='viridis')
plt.colorbar(contour)
plt.title(r'Contour Plot of $(1-x)^2 + 100(y-x^2)^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
